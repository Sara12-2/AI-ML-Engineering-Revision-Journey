### 1. Scalars , Vectors , matrices & Tensors

### 1.1 Scalars - Single Values

A scalar is a single number that represents a magnitude without direction. In machine learning, scalars are used for hyperparameters, metrics, and single values.

```python
import numpy as np

# ML scalars examples
learning_rate = 0.01
accuracy = 0.95
loss = 0.234
epochs = 100

print(f"Learning Rate: {learning_rate}")
print(f"Accuracy: {accuracy}")
print(f"Loss: {loss}")
print(f"Epochs: {epochs}")
print(f"Data Type: {type(accuracy)}")
```

### 1.2 Vectors - 1D Arrays

Vectors are ordered collections of numbers used to represent features, embeddings, and weights in machine learning models.

```python
import numpy as np
# Creating vectors for ML
features = np.array([2.5, 3.7, 1.8, 0.5])
weights = np.array([0.2, -0.1, 0.5, 0.3])
embedding = np.array([0.1, -0.2, 0.3, 0.4])

print(f"Feature Vector: {features}")
print(f"Shape: {features.shape}")
print(f"Weight Vector: {weights}")
print(f"Word Embedding: {embedding}")
```

### 1.3 Matrices - 2D Arrays

Matrices are rectangular arrays of numbers arranged in rows and columns, used to represent datasets and model weights in neural networks.

```python
# Dataset matrix (4 samples, 3 features)
dataset = np.array([
    [2.5, 3.7, 1.8],
    [1.2, 4.1, 2.3],
    [3.3, 2.9, 1.5],
    [2.8, 3.2, 2.1]
])

# Weight matrix for neural network layer
weights_matrix = np.array([
    [0.2, -0.1, 0.5],
    [0.3, 0.4, -0.2]
])

print("Dataset (4x3):")
print(dataset)
print(f"\nShape: {dataset.shape}")
print("\nWeights Matrix (2x3):")
print(weights_matrix)
```

### 1.4 Tensors - Multi-Dimensional Arrays

Tensors are multi-dimensional arrays that extend beyond 2D matrices. In deep learning, tensors represent complex data like images, videos, and batches.

```python
# 3D Tensor: Batch of grayscale images (batch, height, width)
grayscale_batch = np.random.rand(32, 28, 28)
print("3D Tensor (Grayscale Images):")
print(f"Shape: {grayscale_batch.shape}")

# 4D Tensor: Batch of RGB images (batch, height, width, channels)
rgb_batch = np.random.rand(16, 224, 224, 3)
print("\n4D Tensor (RGB Images):")
print(f"Shape: {rgb_batch.shape}")

# 5D Tensor: Batch of videos (batch, frames, height, width, channels)
video_batch = np.random.rand(8, 30, 64, 64, 3)
print("\n5D Tensor (Video Frames):")
print(f"Shape: {video_batch.shape}")
```
